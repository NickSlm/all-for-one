```python
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from sklearn.metrics import confusion_matrix
import kagglehub
import matplotlib.pyplot as plt
import cv2
from keras.callbacks import TensorBoard
```

```python
DATASET_DIR = "/home/nick/ml1/datasets/date-fruit-image-dataset-in-controlled-environment/versions/1"
MODELS_DIR = "/home/nick/all-for-one/backend/models/models"
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
    DATASET_DIR,
    validation_split=0.3,
    subset="training",
    seed=123,
    image_size=(224, 224),
    batch_size=32
)
val_ds = tf.keras.utils.image_dataset_from_directory(
    DATASET_DIR,
    validation_split=0.3,
    subset="validation",
    seed=123,
    image_size=(224, 224),
    batch_size=32
)

test_split = 0.5
val_batches = tf.data.experimental.cardinality(val_ds)
test_size = val_batches // 2

test_ds = val_ds.take(test_size)
val_ds = val_ds.skip(test_size)

class_names = train_ds.class_names
```

```python
def normalize_image(image, label):
    image = tf.cast(image, tf.float32) / 255.0
    return image, label
```

```python
train_ds = train_ds.map(lambda x, y: normalize_image(x, y))
test_ds = test_ds.map(lambda x, y: normalize_image(x, y))
val_ds = val_ds.map(lambda x, y: normalize_image(x, y))
```

```python
n_outputs = len(class_names)

tensorboard_cb = TensorBoard(log_dir="./logs", histogram_freq=0, write_graph=True, write_images=False)

model = keras.models.Sequential([
    keras.layers.Conv2D(32, (3,3), activation="relu", input_shape=(224,224,3)),
    keras.layers.MaxPool2D((2, 2)),
    keras.layers.Conv2D(64, (3,3), activation="relu"),
    keras.layers.MaxPool2D((2, 2)),
    keras.layers.Conv2D(128, (3,3), activation="relu"),
    keras.layers.MaxPool2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(256, activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(n_outputs, activation="softmax")
])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
history = model.fit(train_ds, epochs=100, validation_data=test_ds, callbacks=[tensorboard_cb])

file_name = "imageClassification.h5"

model.save(os.path.join(MODELS_DIR, file_name))
```

```python
steps = np.arange(len(history.history["accuracy"]))
plt.plot(steps, history.history["accuracy"])
plt.grid()
```

```python
val_images = []
val_labels = []
for images, labels in val_ds:
    val_images.append(images.numpy())
    val_labels.append(labels.numpy())
```

```python
val_images = np.concatenate(val_images)
val_labels = np.concatenate(val_labels)
```

```python
pred_labels = model.predict(val_images)
pred_labels = np.argmax(pred_labels, axis=1)
```

```python
cm = confusion_matrix(val_labels, pred_labels, labels=np.arange(len(class_names)))
```

```python
import seaborn
```

```python
figure = plt.figure(figsize=(8, 6))
seaborn.heatmap(cm, annot=True, fmt="d", cmap="Blues",xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.xticks(rotation=30)
plt.yticks(rotation=30)  # Rotate y-axis labels by 30 degrees
plt.show()
```

```python
image_name = 'Medjool-Dates.jpg'
image_path = os.path.join("/home/nick/all-for-one/backend/models", image_name)
```

```python
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_resized = cv2.resize(image_rgb, (224, 224), interpolation=cv2.INTER_AREA)
    image_resized = image_resized / 255.0
    
    image_resized = np.expand_dims(image_resized, axis=0)
    return image_resized
```

```python
image = preprocess_image(image_path)
label_pred = model.predict(image)
label_pred = np.argmax(label_pred)
```

```python
figure = plt.figure(figsize=(3,3))
plt.imshow(image_rgb)
plt.title(class_names[label_pred])
plt.axis("off")
```
